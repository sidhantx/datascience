```python
def create_dir_with_data_for_tf(orig_dir, base_dir):
    """
    This method preprocess the training, val, and test data in to
    a directory structure that is used by tensorflow to convert the 
    images to tensorflow data. It uses the provided truth.txt to 
    create this structure.
    
    params : 
            orig_dir : 100x100 dir provided with the assignment
            base_dir : new folder called data which stores our
            data in to train, test and val data with each folder
            having 3 sub-folders namely abstract, landscape, and 
            portrait. The names corresponds to the labels of our
            data that we are classifying.
            
    returns: returns a list with names of files missing i.e. the file
             exists in truth.txt but does not exist in the provided
             data. This is for reference purpose only. 
    
    
    """
    files_missing = []
    
    for directory in ('train','validate', 'test'):
        for cl in ('abstract', 'landscape', 'portrait'):
            new_dir = base_dir / directory / cl
            os.makedirs(new_dir)
            
    with open("truth.txt") as truth:
        i=0
        for line in truth:
            if i > 0:
                split = line.split('\t')
                
                if os.path.isfile(orig_dir /split[1]/split[0]):
                
                    if  os.path.isdir(base_dir/split[1]/split[2]):
                
                        shutil.copyfile(src = orig_dir /split[1]/split[0], \
                            dst = base_dir/split[1]/split[2]/split[0])
            
                    elif split[2] == "self-portrait":
                        split[2] = "portrait"
                        shutil.copyfile(src = orig_dir /split[1]/split[0], \
                            dst = base_dir/split[1]/split[2]/split[0])
                    
                else:
                    files_missing.append(split)
        
            i += 1
    
    return files_missing
    
       
```

```python
orig_dir = pathlib.Path("100x100")
base_dir = pathlib.Path("data")

missing_files = []
if not os.path.isdir(base_dir):
    missing_files = create_dir_with_data_for_tf(orig_dir, base_dir)
    
print(f"Files in Truth.txt but not in data :\n")
print(missing_files)
  
```

**Other helper methods used in the assignment for plots, and callsbacks.**

```python
def plot_model_training(train_history, suptitle,):
    """
    Helper method to plot train and val accuracy as a function of
    epochs, and loss as a function of epochs
    
    params:
        train_history : training history
        suptitle : suptitle for both the plots
    """

    n_epoch = np.argmin(train_history.history['val_loss'])

    loss_train = train_history.history["loss"][n_epoch]
    loss_val = train_history.history["val_loss"][n_epoch]
    acc_train = train_history.history["accuracy"][n_epoch]
    acc_val = train_history.history["val_accuracy"][n_epoch]


    best_epoch = np.argmax(np.array(train_history.history['val_accuracy']))


    fig, ax = plt.subplots(1,2,figsize = (16,6))

    x_axis_vals = np.array(range(len(train_history.history["accuracy"])))+1

    fig.suptitle(
    suptitle.format(
        len(x_axis_vals)
    ),
    fontsize=20,
    )

    ax[0].plot(
    x_axis_vals,
    train_history.history["accuracy"],
    "--",
    label="train accuracy",
    )
    ax[0].plot(
    x_axis_vals,
    train_history.history["val_accuracy"],
    "-",
    label="val accuracy",
    )
    
    ax[0].axvline(
    n_epoch+1,
    color="k",
    linestyle=":",
    label=f'best val accuracy = {acc_val:.2f}'
    )

    ax[0].set_title("Accuracy as a function of epochs ")

    ax[1].plot(
    x_axis_vals,
    train_history.history["loss"],
    "--",
    label="train loss",
    )

    ax[1].plot(
    x_axis_vals,
    train_history.history["val_loss"],
    "-",
    label="val loss",
    )

    ax[1].axvline(
    n_epoch+1,
    color="k",
    linestyle=":",
    label=f'best val loss = {loss_val:.2f}'
    )

    ax[1].set_title("loss as a function of epochs ")

    ax[0].tick_params(labelsize=12)
    ax[0].set_ylabel("accuracy", fontsize=12)
    ax[0].set_xlabel("epoch", fontsize=12)
    ax[0].grid(':', alpha=0.4)
    ax[0].legend(fontsize=12)
        
    ax[1].grid(':', alpha=0.4)
    ax[1].tick_params(labelsize=12)
    ax[1].legend(fontsize=12)    
    ax[1].set_xlabel('epoch')
    ax[1].set_ylabel('loss')

    plt.tight_layout()
    plt.show();


def model_callbacks(filepath, patience =8):
    """
    This method returns the list of callbacks \
    used in model training. 
    Callbacks used :
                    ModelChecktpoint
                    EarlyStopping
    
    params:
            filepath : path to save the weights
            patience : epochs to wait before early \
                        stopping. defalut = 8
    
    returns: List with ModelCheckpoint, and EarlyStopping callback
    """
    callbacks = [
    keras.callbacks.ModelCheckpoint(
    filepath=filepath,
    save_best_only=True,
    monitor="val_loss"),
    
    tf.keras.callbacks.EarlyStopping(
        monitor="val_loss",
        min_delta=0,
        verbose=1, 
        patience= patience,
        restore_best_weights= True
    ) ,
        
    tf.keras.callbacks.CSVLogger(
    filename = f"{filepath}/training_history.log",
    separator = ',',
    append = False
    
    )
    ]
    
    return callbacks


def plot_helper_multiple(train_history, ax1,ax2, title):
    """
     Helper method to plot train and val accuracy as a function of
     epochs, and loss as a function of epochs on an axis
    
     params:
        train_history : training history
        ax1: axis for plotting accuracy as a function of epoch
        ax2: axis for plotting loss as a function of epoch
        base_filter : base filter used in the CNN model
    """
    
    n_epoch = np.argmin(train_history.history['val_loss'])

    loss_train = train_history.history["loss"][n_epoch]
    loss_val = train_history.history["val_loss"][n_epoch]
    acc_train = train_history.history["accuracy"][n_epoch]
    acc_val = train_history.history["val_accuracy"][n_epoch]


    best_epoch = np.argmax(np.array(train_history.history['val_accuracy']))

    x_axis_vals = np.array(range(len(train_history.history["accuracy"])))+1

    #fig.suptitle(suptitle + str(base_filter),fontsize=14)

    ax1.plot(
    x_axis_vals,
    train_history.history["accuracy"],
    "--",
    label="train accuracy",
    )
    ax1.plot(
    x_axis_vals,
    train_history.history["val_accuracy"],
    "-",
    label="val accuracy",
    )
    
    ax1.axvline(
    n_epoch+1,
    color="k",
    linestyle=":",
    label=f'best val accuracy = {acc_val:.2f}'
    )

    ax1.set_title(
        f"Accuracy as a function of epochs {title}",
        fontsize = 18,  
        weight='bold')
    
    ax1.tick_params(labelsize=12)
    ax1.set_ylabel("accuracy", fontsize=12)
    ax1.set_xlabel("epoch", fontsize=12)
    ax1.grid(':', alpha=0.4)
    ax1.legend(fontsize=12)

    ax2.plot(
    x_axis_vals,
    train_history.history["loss"],
    "--",
    label="train loss",
    )

    ax2.plot(
    x_axis_vals,
    train_history.history["val_loss"],
    "-",
    label="val loss",
    )

    ax2.axvline(
    n_epoch+1,
    color="k",
    linestyle=":",
    label=f'best val loss = {loss_val:.2f}'
    )

    ax2.set_title(
        f"loss as a function of epochs {title})",
        fontsize=18,
        weight='bold')
    
    ax2.grid(':', alpha=0.4)
    ax2.tick_params(labelsize=12)
    ax2.legend(fontsize=12)    
    ax2.set_xlabel('epoch')
    ax2.set_ylabel('loss')
```

## Baseline CNN Model 

<p><b>
For my baseline CNN model, I am using the VGG16 architecture as described in the paper
    <a>https://arxiv.org/pdf/1409.1556v6.pdf</a>
    
I also reference this website 
https://www.mygreatlearning.com/blog/introduction-to-vgg16/
    
I created the architecture from scratch , but since the link https://www.mygreatlearning.com/blog/introduction-to-vgg16/ implements the same architecture some code may look similar. 

I wanted to experiment with number of filters, regularizers, dropout rate, dense layer nodes and strides.These are used as parameters to the function that creates this model.  

My baseline uses a small filter base of 2, giving me 2 filters in first block, 4 in second block , 8 in 3rd block and 16 filters in block 4 and 5. I use max pool size of 2X2 to reduce the size of the images as dicussed in class and the above paper. For the baseline model I am not using any strides.  

<b></p>

```python
def create_cnn_model(filter_base = 2, \
                     kernel_regularizer = None,\
                     kernel_regularizer_dense =None,\
                     dropout_rate =0, pool_size = (2,2),\
                     strides = None,\
                     dense_nodes =256,
                     use_block4 = True,
                     use_block5 = True
                    ):
    """
    This method creates a model using VGG16 architecture. The default
    params use less number of filter in each Convolution layer. The 
    dense layers use less nodes as default. 
    """
    
    model =Sequential()
    
    #rescale the data
    model.add(layers.Rescaling(1./255, input_shape=(100, 100, 3)))
    
    #Block 1
    model.add(Conv2D(filter_base, kernel_size =3, padding = 'same',\
                        kernel_regularizer = kernel_regularizer))
    
    model.add(Activation('relu'))
    
    model.add(Conv2D(filter_base, kernel_size= 3, padding='same',\
                    kernel_regularizer= kernel_regularizer))
    model.add(Activation('relu'))

    model.add(MaxPooling2D(pool_size= pool_size, strides= strides))
    
              
    #Block 2 layers
    model.add(Conv2D(filter_base*2, kernel_size =3, padding = 'same',\
                        kernel_regularizer = kernel_regularizer))
    
    model.add(Activation('relu'))
    
    model.add(Conv2D(filter_base*2, kernel_size= 3, padding='same',\
                    kernel_regularizer= kernel_regularizer))
    model.add(Activation('relu'))

    model.add(MaxPooling2D(pool_size= pool_size, strides= strides))
    
    
    # Block 3 layers
    model.add(Conv2D(filter_base * 4, kernel_size= 3, padding='same',\
                kernel_regularizer= kernel_regularizer))
    model.add(Activation('relu'))

    model.add(Conv2D(filter_base * 4, kernel_size= 3, padding='same',\
    kernel_regularizer= kernel_regularizer))
    model.add(Activation('relu'))
    
    model.add(MaxPooling2D(pool_size= pool_size, strides= strides))
    
    if use_block4:
        # Block 4 layers
        model.add(Conv2D(filter_base * 8, kernel_size= 3, padding='same',\
                    kernel_regularizer= kernel_regularizer))
        model.add(Activation('relu'))

        model.add(Conv2D(filter_base * 8, kernel_size= 3, padding='same',\
                        kernel_regularizer = kernel_regularizer))
        model.add(Activation('relu'))
    
        model.add(MaxPooling2D(pool_size= pool_size, strides= strides))
    
    if use_block5:          
        # Block 5 layers
        model.add(Conv2D(filter_base * 8, kernel_size= 3, padding='same',\
                        kernel_regularizer= kernel_regularizer))
        model.add(Activation('relu'))

        model.add(Conv2D(filter_base * 8, kernel_size= 3, padding='same',\
                        kernel_regularizer = kernel_regularizer))
        model.add(Activation('relu'))
    
        model.add(MaxPooling2D(pool_size= pool_size, strides= strides))
    
    
              
    #Flatten
    model.add(Flatten())
    
    #Add dense layers          
    model.add(Dense(dense_nodes, activation ='relu', \
                    kernel_regularizer= kernel_regularizer_dense))
    model.add(Dropout(dropout_rate))
    
    model.add(Dense(dense_nodes, activation ='relu', \
                    kernel_regularizer= kernel_regularizer_dense))
    model.add(Dropout(dropout_rate))
    
    #output layer
    model.add(Dense(3, activation = 'softmax'))
    
    return model
    
```

```python

baseline_cnn_model = create_cnn_model()

cnn_callbacks = model_callbacks("checkpoints/base_cnn",30)


baseline_cnn_model.compile(
    optimizer = optimizers.Adam(),
    loss = 'categorical_crossentropy',
    metrics = ['accuracy']

)

print(baseline_cnn_model.summary())
```

```python
model_list_improved = []

base_filters = [16, 32,64]
block_4_layers = [True]
block_5_layers = [False]
i=0
for base_filter in base_filters:
    for block4 in block_4_layers:
        for block5 in block_5_layers:
            
            #clear session
            tf.keras.backend.clear_session();
            gc.collect()
            
            params = (base_filter, block4, block5)
            print(f"Running Model {i+1} with params {params}")
            
            cnn_model = create_cnn_model(
                            filter_base= base_filter,
                            use_block4 = block4,
                            use_block5 = block5)
    
            checkpoint_path = f"checkpoints/p3/{i}"
    

            cnn_callbacks = model_callbacks(
                checkpoint_path, 
                patience= 10)

            cnn_model.compile(
                optimizer = optimizers.Adam(),
                loss = 'categorical_crossentropy',
                metrics = ['accuracy']
                )
    
            history_cnn = cnn_model.fit(
            train_data,
            epochs = 100,
            validation_data = validation_data,
            callbacks = cnn_callbacks
            )
            model_list_improved.append((cnn_model, history_cnn, params))
            i = i+1
    
```

## Problem 5 Optimized CNN classifier -- 20 points

### Further optimize the classifier to achieve the best results that you can.   Create a plot of the training and validation results as a function of the epoch.   Which result gave the best results?

<p><b>
The optimized model uses data augmentation as first layer. It also has only 4 blocks compared to the 
    5 blocks each with 2 Convolution layers. Thus decreasing some complexity. 
<b><p>

```python
def create_optimized_cnn_model(filter_base,data_augmentation,\
                              dropout_rate =0.2,
                              dense_nodes =256):

    model = Sequential([
                data_augmentation,
                layers.Rescaling(1./255),
    
                #Block1 
                layers.Conv2D(filter_base, kernel_size =3, padding = 'same',\
                        activation ='relu'),
                layers.Conv2D(filter_base, kernel_size =3, padding = 'same',\
                        activation ='relu'),
                layers.MaxPooling2D(),
    
                #Block2
        
                layers.Conv2D(filter_base*2, kernel_size =3, padding = 'same',\
                        activation ='relu'),
                layers.Conv2D(filter_base, kernel_size =3, padding = 'same',\
                        activation ='relu'),
                layers.MaxPooling2D(),
        
                #Block 3
                layers.Conv2D(filter_base*4, kernel_size =3, padding = 'same',\
                        activation ='relu'),
                layers.Conv2D(filter_base, kernel_size =3, padding = 'same',\
                        activation ='relu'),
                layers.MaxPooling2D(),
        
                #Block 4
                layers.Conv2D(filter_base*8, kernel_size =3, padding = 'same',\
                        activation ='relu'),
                layers.Conv2D(filter_base, kernel_size =3, padding = 'same',\
                        activation ='relu'),
                layers.MaxPooling2D(),
                    
                layers.Flatten(),
                layers.Dense(dense_nodes, activation ='relu'),
                layers.Dropout(dropout_rate),
                layers.Dense(dense_nodes/2,activation ='relu'),
                layers.Dropout(dropout_rate),
                layers.Dense(3, activation = 'softmax')
        
                ])
    
    return model
```

**Model training - grid search on dropout rate with learning rate of 1e-4. Default on Adam is 1e-3**

```python
data_augmentation = keras.Sequential(
  [
    layers.RandomFlip("horizontal",
                      input_shape=(100,
                                  100,
                                  3)),
    layers.RandomRotation(0.1),
    layers.RandomZoom(0.1),
  ]
)


model_list = []

dropout_rates =[0.2,0.3,0.4, 0.5]
learning_rates =[1e-4]

i=0
for lr in learning_rates:
    for rate in dropout_rates:
        tf.keras.backend.clear_session();
        gc.collect()
        params = (lr, rate, 32, 256)
        print(f"Running model no. {i} with{params}")
        
        cnn_model = create_optimized_cnn_model(filter_base= 32,
                                      data_augmentation= data_augmentation,
                                      dropout_rate= rate)
        
        cnn_callbacks = model_callbacks(
                                f"checkpoints/p5/cnn_{i}", patience= 15)

        cnn_model.compile(optimizer = optimizers.Adam(learning_rate= lr),
                            loss = 'categorical_crossentropy',
                            metrics = ['accuracy']
                            )
    
        history_cnn = cnn_model.fit(train_data,
                            epochs = 300,
                            validation_data = validation_data,
                            callbacks = cnn_callbacks
                             )
        model_list.append((cnn_model, history_cnn, params))
        i = i+1
               
                
```

```python
def calculate_show_accuracy(tp,fp,label):
    """
    Calculate accuracy and prints accuracy 
    params : tp : True Positive for a class or all
             fp : False Positive for a class or all
    """
    accuracy =  (tp)/(tp+fp)
    s = f"\nAccuracy of our model in predicting {label} images"
    s += f" is {accuracy:.4f}"
    print(s)
```

**Show misclassified images**

```python
def get_misclassified_images(number_of_misclassified,\
                             test_data,\
                             model):
    """
    helper method to get misclassified image and their true and 
    predicted labels.
    """
    classes  =[0,1,2]
    class_count = [0,0,0]
    images_to_show = number_of_misclassified
    image_list =[]
    for x, y in test_data:
        predictions = model.predict(x)
        for i in range(predictions.shape[0]):
            y_actual = np.argmax(y[i])
            score = tf.nn.softmax(predictions [i])
            y_pred = np.argmax(score)
        
            if y_actual != y_pred:
                if y_actual == 0 and class_count[0] <=4:
                    image_list.append((x[i], y_actual, y_pred))
                    class_count[0] = class_count[0] +1 
                elif  y_actual == 1 and class_count[1] <=4:
                    image_list.append((x[i], y_actual, y_pred))
                    class_count[1] = class_count[1] +1 
            
                elif  y_actual == 2 and class_count[2] <=4:
                    image_list.append((x[i], y_actual, y_pred))
                    class_count[2] = class_count[2] +1 
        
        
            if len(image_list) == images_to_show:
                break
        if len(image_list) >=  images_to_show:
            break
    return image_list

```

height of the image as it moves through the network. 

However in Feed Forward neural network is limited because we have each neuron with weights and
a activation function. It does not have the ability of CNN for applying filters to learn edges.

The above mentioned aspects of CNN architecture contributed to giving better performance on
on our classification of images.
